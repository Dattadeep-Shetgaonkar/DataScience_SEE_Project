```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --------------------------------------------------------
#  LOAD DATA
# --------------------------------------------------------
df = pd.read_csv("healthcare_dataset_cleaned.csv")
print("âœ… Dataset loaded successfully:", df.shape)

# Convert date columns to datetime (important for Length of Stay)
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"], errors='coerce')
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"], errors='coerce')

# --------------------------------------------------------
# 1. BASIC STATISTICS (single clean table)
# --------------------------------------------------------
print("\nðŸ“Š BASIC STATISTICS:")
print(df.describe(include='all').to_string())

# --------------------------------------------------------
# 2. AGE DISTRIBUTION
# --------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], kde=True, bins=20, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --------------------------------------------------------
# 3. MEDICAL CONDITION COUNT
# --------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.countplot(
    y="Medical Condition",
    data=df,
    order=df["Medical Condition"].value_counts().index,
    palette="viridis"
)
plt.title("Medical Condition Frequency")
plt.xlabel("Count")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.show()

# --------------------------------------------------------
# 4. BLOOD TYPE DISTRIBUTION
# --------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.countplot(
    x="Blood Type",
    data=df,
    order=df["Blood Type"].value_counts().index,
    palette="plasma"
)
plt.title("Blood Type Distribution")
plt.xlabel("Blood Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --------------------------------------------------------
# 5. ADMISSION TYPE DISTRIBUTION
# --------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.countplot(
    x="Admission Type",
    data=df,
    order=df["Admission Type"].value_counts().index,
    palette="coolwarm"
)
plt.title("Admission Type Distribution")
plt.xlabel("Admission Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --------------------------------------------------------
# 6. LENGTH OF STAY ANALYSIS
# --------------------------------------------------------

# Create Length of Stay column
df["Length of Stay"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days

print("\nðŸ•’ LENGTH OF STAY CREATED:")
print(df[["Date of Admission", "Discharge Date", "Length of Stay"]].head())



# --------------------------------------------------------
# 7. CORRELATION HEATMAP 
# --------------------------------------------------------
numeric_df = df.select_dtypes(include=["float64", "int64"])

plt.figure(figsize=(8, 6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    square=True  # keeps it visually 4x4
)
plt.title("Correlation Heatmap (Numeric Features)")
plt.tight_layout()
plt.show()

# --------------------------------------------------------
# 8. COMPLETION
# --------------------------------------------------------
print("\nâœ… EDA Completed Successfully !")
```
